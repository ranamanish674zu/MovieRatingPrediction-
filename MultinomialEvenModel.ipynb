```python
#training data
x=["This was an awesome movie",
   "Great movie! i liked it a lot ",
   "Happy Ending ! awesome acting by the hero",
   "Loved it ! trully great",
   "bad not upto the mark",
   "could have been better ",
   "surely a Disappointing movie"]
#testing data
y=[1,1,1,1,0,0,0]# 1-positive ,0-negative class 
```

```python
test_x =["I was happy & happy and I loved the acting in the movie","The movie I saw was  bad"]
```

Cleaning

```python
#cleaning functioon the training and testing data
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
def getClenaReview(review):
    review = review.lower()
    review = review.replace("<br/><br/>", " ")
    # Tokenize
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(review)
    # Remove stopwords
    en_stopwords = set(stopwords.words('english'))
    # my_stopwords = set(["http", "'s", "n't", "'m", "'re", "'ve"])
    # en_stopwords.update(my_stopwords)
    new_tokens = [token for token in tokens if token not in en_stopwords]
    # Stem tokens
    ps = PorterStemmer()
    stemmed_tokens = [ps.stem(token) for token in new_tokens]
    cleaned_review = ' '.join(stemmed_tokens)
    return cleaned_review
```

```python
#cleaning  the training and testing data
x_clean=[getClenaReview(i) for i in x]#list comprahension
xt_clean=[getClenaReview(i)for i in test_x]
print(x_clean)
print(xt_clean)
```

Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
#converting vocab into the vector 
x_vec=cv.fit_transform(x_clean).toarray()
print(x_vec)
print(x_vec.shape)
#each row represent the document from vacab or clean data.
#column represent the no of uniqur words comes 
#here we are taking the document from the vocab or clean_data and chake the word form the clean data in which index lying in the features_names.then in
#which index the word from clean data is present in features name that index will be assined as 1 or others can be assign as 0 in the matrix .
```

```python
cv = CountVectorizer()
x_vec = cv.fit_transform(x_clean).toarray()

feature_names = cv.get_feature_names_out()
print(feature_names)

```

```python
#vectorization on the test set 
xt_vec=cv.transform(xt_clean).toarray()
print(xt_vec)
cv.get_feature_names_out()
print(xt_vec.shape)
```

 Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,GaussianNB
```

```python
mnb=MultinomialNB()
print(mnb)
```

```python
#Training 
mnb.fit(x_vec,y)
```

```python
#predictions
mnb.predict(xt_vec)
#it's predicting the first review is positive and second is negative
```

```python
#predictiong probability
mnb.predict_proba(xt_vec)
```

```python
mnb.score(x_vec,y)#accuracy
```
